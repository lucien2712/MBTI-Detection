##Stopwords removel

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words("english"))
```

```python
import string

def stopword_removel(x):
  new_x = []
  for sent in x:
    c = " ".join([i for i in sent.split() if i not in stop_words])
    new_x.append(c)
  return new_x

data["posts"] = data["posts"].apply(stopword_removel)
data["posts"][0]
```

##Punctuations removel

```python
import string

def punctuations_removel(x):
  new_x = []
  for word in x:
    a = "".join([alpha for alpha in word if alpha not in string.punctuation])
    if a != "":
      new_x.append(a)  
  return new_x
data["posts"] = data["posts"].apply(punctuations_removel)
data["posts"][0]
```

## special token for Links 

```python
import re

def links_token(x):
  result = []
  for word in x:
    filtered_sentence = re.sub(r'https?\S+', ' [Link] ', word)
    result.append(filtered_sentence)
  return result

data["posts"] = data["posts"].apply(links_token)
data["posts"][0]
```

##Numbers removel

```python
def numbers_token(x):
  result = []
  for word in x:
    filtered_sentence = re.sub(r'\d+', ' ', word)
    result.append(filtered_sentence)
  return result
      
  
data["posts"] = data["posts"].apply(numbers_token)
data["posts"][0]
```

## **other preprocessing**

```python
def tidy(x):
  result = []
  for i in x:
    a = i.strip()
    result.append(a)
  return result
data["posts"] = data["posts"].apply(tidy)  #刪除詞多於留白(" hello ")
data["posts"][0]
```

```python
def blank(x):
  result = []
  for i in x:
    if len(i) > 0:
      result.append(i)
  return result
data["posts"] = data["posts"].apply(blank)   #刪除空白("")
data["posts"][0]
```

```python
import torch
import torch.nn as nn
from transformers import BertModel,RobertaModel
```

```python
device = "cuda" if torch.cuda.is_available()  else "cpu"
print(device)
```

```python
class multimodel(nn.Module):
    def __init__(self):
        super(multimodel, self).__init__()
        self.bert = BertModel.from_pretrained('bert-base-uncased')
        #self.bert = RobertaModel.from_pretrained('roberta-base')
        self.bert.resize_token_embeddings(len(tokenizer))
        
        
        #--------------------------------------------------------------------------------------->主要更改區域
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=(3,3)) 
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=(3,3))
        #self.conv3 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=(3,1))                                     
        self.maxpool = nn.MaxPool2d((3,3))
        self.conv_output = 32*12*254   #our output size
        #self.conv_output = 32*36*764   #output shape of other structure

        self.dense_layer_1 = nn.Linear(self.conv_output + 64, 128) 
        self.dense_layer_2 = nn.Linear(128, 4)
        self.act_sig = nn.Sigmoid()
        #--------------------------------------------------------------------------------------->主要更改區域
    def forward(self, input_ids, attention_mask,liwc):

        cls_tensor = None

        for i in range(len(input_ids)):
          cls = self.bert(input_ids[i], attention_mask = attention_mask[i])[0][:,0,:].unsqueeze(0) #1 * 40 * 768
          #print("cls : ", cls.shape)

          if i == 0:
            cls_tensor = cls.clone()
            
          else:
            cls_tensor = torch.cat([cls_tensor, cls], dim = 0) #batch_size * 40 * 768
            #print("cls tensor: ", cls_tensor.shape)
     
        cls_tensor = cls_tensor.unsqueeze(1) # batch_size*1*40*768
        #print("cls tensor: ", cls_tensor.shape)

        out = self.conv1(cls_tensor) 
        out = self.conv2(out)
        #out = self.conv3(out)
        out = self.maxpool(out)
 
        #print("conv1 out shape:", out.shape)

        out = out.view(out.shape[0], 1, self.conv_output)
        out = torch.cat([out, liwc.unsqueeze(1)],dim=2)
        
        out = self.dense_layer_1(out)
        out = self.dense_layer_2(out)

        out = self.act_sig(out)

    
        return out


```

```python
from torch import optim
device = "cuda" if torch.cuda.is_available()  else "cpu"
print(device)

device=torch.device(device)
model = multimodel() 
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.00002) #-----------------------------------------------------主要更改區域
#scheduler = optim.lr_scheduler.StepLR(optimizer, step_size = 2, gamma = 0.1)

model = model.to(device)
```
